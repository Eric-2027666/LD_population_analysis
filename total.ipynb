```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
import numpy as np
```

```python
import os
country_list = os.listdir("./1_data")
country_codes = {'AUS': 'Australia',
                 'AUT': 'Austria',
                 'BEL': 'Belgium',
                 'BGR': 'Bulgaria',
                 'BLR': 'Belarus',
                 'CAN': 'Canada',
                 'CHE': 'Switzerland',
                 'CHL': 'Chile',
                 'CZE': 'Czech Republic',
                 'DEUTE': 'Germany',
                 'DEUTNP': 'Germany',
                 'DEUTW': 'Germany',
                 'DNK': 'Denmark',
                 'ESP': 'Spain',
                 'EST': 'Estonia',
                 'FIN': 'Finland',
                 'FRACNP': 'France',
                 'FRATNP': 'France',
                 'GBRCENW': 'Great Britain',
                 'GBRTENW': 'Great Britain',
                 'GBR_NIR': 'Great Britain',
                 'GBR_NP': 'Great Britain',
                 'GBR_SCO': 'Great Britain',
                 'GRC': 'Greece',
                 'HKG': 'Hong Kong',
                 'HRV': 'Croatia',
                 'HUN': 'Hungary',
                 'IRL': 'Ireland',
                 'ISL': 'Iceland',
                 'ISR': 'Israel',
                 'ITA': 'Italy',
                 'JPN': 'Japan',
                 'KOR': 'South Korea',
                 'LTU': 'Lithuania',
                 'LUX': 'Luxembourg',
                 'LVA': 'Latvia',
                 'NLD': 'Netherlands',
                 'NOR': 'Norway',
                 'NZL_MA': 'New Zealand',
                 'NZL_NM': 'New Zealand',
                 'NZL_NP': 'New Zealand',
                 'POL': 'Poland',
                 'PRT': 'Portugal',
                 'RUS': 'Russia',
                 'SVK': 'Slovakia',
                 'SVN': 'Slovenia',
                 'SWE': 'Sweden',
                 'TWN': 'Taiwan',
                 'UKR': 'Ukraine',
                 'USA': 'United States'}
```

```python
double_index_total = pd.DataFrame()
single_index_total = pd.DataFrame()
for country in country_list:
    country_dir = country
    
    # double index
    population = pd.read_csv("./1_data/" + country_dir + "/STATS/" + "Population.txt", sep = "\s+", header = 1)
    population['Year'] = population['Year'].astype('str')
    has_minus = population['Year'].str.contains('-')
    no_minus = ~has_minus
    population = population[no_minus].reset_index(drop=True)
    population['Year'] = population['Year'].str.replace('[\+\-]', '', regex=True)
    population['Age'] = population['Age'].astype('str')
    
    bltper = pd.read_csv("./1_data/" + country_dir + "/STATS/" + "bltper_1x1.txt", sep = "\s+", header = 1)
    bltper.rename(columns={'mx':'Central_death_rate_total', 'qx':'Probability_of_death_total', 'ax':'Average_length_of_survival_total', 'lx':'Number_of_survivors_total', 'dx':'Number_of_deaths_total', 'Lx':'Number_of_person-years_lived_total', 'Tx':'Number of person-years remaining_total', 'ex':'Life_expectancy_total'},inplace=True)
    bltper['Year'] = bltper['Year'].astype('str')
    has_minus = bltper['Year'].str.contains('-')
    no_minus = ~has_minus
    bltper = bltper[no_minus].reset_index(drop=True)
    bltper['Year'] = bltper['Year'].str.replace('[\+\-]', '', regex=True)
    bltper['Age'] = bltper['Age'].astype('str')
    two_index = pd.merge(population, bltper, on=['Year', 'Age'], how='outer')
    
    fltper = pd.read_csv("./1_data/" + country_dir + "/STATS/" + "fltper_1x1.txt", sep = "\s+", header = 1)
    fltper.rename(columns={'mx':'Central_death_rate_female', 'qx':'Probability_of_death_female', 'ax':'Average_length_of_survival_female', 'lx':'Number_of_survivors_female', 'dx':'Number_of_deaths_female', 'Lx':'Number_of_person-years_lived_female', 'Tx':'Number of person-years remaining_female', 'ex':'Life_expectancy_female'},inplace=True)
    fltper['Year'] = fltper['Year'].astype('str')
    has_minus = fltper['Year'].str.contains('-')
    no_minus = ~has_minus
    fltper = fltper[no_minus].reset_index(drop=True)
    fltper['Year'] = fltper['Year'].str.replace('[\+\-]', '', regex=True)
    fltper['Age'] = fltper['Age'].astype('str')
    two_index = pd.merge(two_index, fltper, on=['Year', 'Age'], how='outer')
    
    mltper = pd.read_csv("./1_data/" + country_dir + "/STATS/" + "mltper_1x1.txt", sep = "\s+", header = 1)
    mltper.rename(columns={'mx':'Central_death_rate_male', 'qx':'Probability_of_death_male', 'ax':'Average_length_of_survival_male', 'lx':'Number_of_survivors_male', 'dx':'Number_of_deaths_male', 'Lx':'Number_of_person-years_lived_male', 'Tx':'Number of person-years remaining_male', 'ex':'Life_expectancy_male'},inplace=True)
    mltper['Year'] = mltper['Year'].astype('str')
    has_minus = mltper['Year'].str.contains('-')
    no_minus = ~has_minus
    mltper = mltper[no_minus].reset_index(drop=True)
    mltper['Year'] = mltper['Year'].str.replace('[\+\-]', '', regex=True)
    mltper['Age'] = mltper['Age'].astype('str')
    two_index = pd.merge(two_index, mltper, on=['Year', 'Age'], how='outer')
    
    death_rates = pd.read_csv("./1_data/" + country_dir + "/STATS/" + "Mx_1x1.txt", sep = "\s+", header = 1)
    death_rates.rename(columns={'Female':'female_death_rate_period', 'Male':'male_death_rate_period', 'Total':'total_death_rate_period'},inplace=True)
    death_rates['Year'] = death_rates['Year'].astype('str')
    has_minus = death_rates['Year'].str.contains('-')
    no_minus = ~has_minus
    death_rates = death_rates[no_minus].reset_index(drop=True)
    death_rates['Year'] = death_rates['Year'].str.replace('[\+\-]', '', regex=True)
    death_rates['Age'] = death_rates['Age'].astype('str')
    two_index = pd.merge(two_index, death_rates, on=['Year', 'Age'], how='outer')
    
    cExposure_path = "./1_data/" + country_dir + "/STATS/" + "cExposures_1x1.txt"
    if os.path.exists(cExposure_path):
        cExposures = pd.read_csv(cExposure_path, sep = "\s+", header = 1)
        cExposures.rename(columns={'Female':'female_exposure_cohort', 'Male':'male_exposure_cohort', 'Total':'total_exposure_cohort'},inplace=True)
        cExposures['Year'] = cExposures['Year'].astype('str')
        has_minus = cExposures['Year'].str.contains('-')
        no_minus = ~has_minus
        cExposures = cExposures[no_minus].reset_index(drop=True)
        cExposures['Year'] = cExposures['Year'].str.replace('[\+\-]', '', regex=True)
        cExposures['Age'] = cExposures['Age'].astype('str')
        two_index = pd.merge(two_index, cExposures, on=['Year', 'Age'], how='outer')
    
    cMx_path = "./1_data/" + country_dir + "/STATS/" + "cMx_1x1.txt"
    if os.path.exists(cMx_path):
        cMx = pd.read_csv(cMx_path, sep = "\s+", header = 1)
        cMx.rename(columns={'Female':'female_death_rate_cohort', 'Male':'male_death_rate_cohort', 'Total':'total_death_rate_cohort'},inplace=True)
        cMx['Year'] = cMx['Year'].astype('str')
        has_minus = cMx['Year'].str.contains('-')
        no_minus = ~has_minus
        cMx = cMx[no_minus].reset_index(drop=True)
        cMx['Year'] = cMx['Year'].str.replace('[\+\-]', '', regex=True)
        cMx['Age'] = cMx['Age'].astype('str')
        two_index = pd.merge(two_index, cMx, on=['Year', 'Age'], how='outer')
    
    deaths = pd.read_csv("./1_data/" + country_dir + "/STATS/" + "Deaths_1x1.txt", sep = "\s+", header = 1)
    deaths.rename(columns={'Female':'female_deaths', 'Male':'male_deaths', 'Total':'total_deaths'},inplace=True)
    deaths['Year'] = deaths['Year'].astype('str')
    has_minus = deaths['Year'].str.contains('-')
    no_minus = ~has_minus
    deaths = deaths[no_minus].reset_index(drop=True)
    deaths['Year'] = deaths['Year'].str.replace('[\+\-]', '', regex=True)
    deaths['Age'] = deaths['Age'].astype('str')
    two_index = pd.merge(two_index, deaths, on=['Year', 'Age'], how='outer')
    
    exposures = pd.read_csv("./1_data/" + country_dir + "/STATS/" + "Exposures_1x1.txt", sep = "\s+", header = 1)
    exposures.rename(columns={'Female':'female_exposure_period', 'Male':'male_exposure_period', 'Total':'total_exposure_period'},inplace=True)
    exposures['Year'] = exposures['Year'].astype('str')
    has_minus = exposures['Year'].str.contains('-')
    no_minus = ~has_minus
    exposures = exposures[no_minus].reset_index(drop=True)
    exposures['Year'] = exposures['Year'].str.replace('[\+\-]', '', regex=True)
    exposures['Age'] = exposures['Age'].astype('str')
    two_index = pd.merge(two_index, exposures, on=['Year', 'Age'], how='outer')
    
    two_index['Year'] = two_index['Year'].str.replace('[\+\-]', '', regex=True)
    #two_index['Country'] = country_dir
    #two_index['Country_name'] = two_index['Country'].map(country_codes)
    two_index['Country'] = country_codes.get(country_dir)
    double_index_total = pd.concat([double_index_total, two_index], axis=0)
    
    # single index
    births = pd.read_csv("./1_data/" + country_dir + "/STATS/" + "Births.txt", sep = "\s+", header = 1)
    births['Year'] = births['Year'].astype('str')
    has_minus = births['Year'].str.contains('-')
    no_minus = ~has_minus
    births = births[no_minus].reset_index(drop=True)
    births['Year'] = births['Year'].str.replace('[\+\-]', '', regex=True)
    
    life_exp_period = pd.read_csv("./1_data/" + country_dir + "/STATS/" + "E0per.txt", sep = "\s+", header = 1)
    life_exp_period.rename(columns={'Female':'female_exp_period', 'Male':'male_exp_period', 'Total':'total_exp_period'},inplace=True)
    life_exp_period['Year'] = life_exp_period['Year'].astype('str')
    has_minus = life_exp_period['Year'].str.contains('-')
    no_minus = ~has_minus
    life_exp_period = life_exp_period[no_minus].reset_index(drop=True)
    life_exp_period['Year'] = life_exp_period['Year'].str.replace('[\+\-]', '', regex=True)
    
    single_index = pd.merge(births, life_exp_period, on=['Year'], how='outer')
    single_index['Year'] = single_index['Year'].str.replace('[\+\-]', '', regex=True)
    
    #single_index['Country'] = country_dir
    single_index['Country'] = country_codes.get(country_dir)
    single_index_total = pd.concat([single_index_total, single_index], axis=0)
```

```python
double_index_total=double_index_total[(double_index_total['Year'].astype('int') >= 1950) & (double_index_total['Year'].astype('int') <= 2020)]
```

```python
double_index_total.to_csv('double_index.csv', index=False)
single_index_total.to_csv('single_index.csv', index=False)
```
